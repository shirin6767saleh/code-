```python
import numpy as np
import matplotlib.pyplot as plt

def create_matrix_C(N):
    """Create matrix C with elements C(k,l) = (2/sqrt(N)) * cos(2*pi*k*l/N)"""
    k = np.arange(N)[:, np.newaxis]
    l = np.arange(N)
    return (2 / np.sqrt(N)) * np.cos(2 * np.pi * k * l / N)

def analyze_all_cases(m_values):
    """Analyze all four N cases and plot their condition numbers"""
    # Setup cases with their respective configurations
    cases = {
        'N=4m': {
            'n_func': lambda m: 4*m,
            'rows': lambda m: (m, 2*m),
            'cols': lambda m: (0, m),
            'color': 'blue',
            'marker': 'o'
        },
        'N=4m+1': {
            'n_func': lambda m: 4*m + 1,
            'rows': lambda m: (m+1, 2*m+1),
            'cols': lambda m: (0, m),
            'color': 'red',
            'marker': 's'
        },
        'N=4m+2': {
            'n_func': lambda m: 4*m + 2,
            'rows': lambda m: (m+1, 2*m+2),
            'cols': lambda m: (0, m+1),
            'color': 'green',
            'marker': '^'
        },
        'N=4m+3': {
            'n_func': lambda m: 4*m + 3,
            'rows': lambda m: (m+1, 2*m+2),
            'cols': lambda m: (0, m+1),
            'color': 'purple',
            'marker': 'D'
        }
    }

    # Initialize results storage
    results = {case: {'cond': []} for case in cases}

    # Compute condition numbers for each case
    for m in m_values:
        for case, config in cases.items():
            N = config['n_func'](m)
            row_start, row_end = config['rows'](m)
            col_start, col_end = config['cols'](m)

            C = create_matrix_C(N)
            B = C[row_start:row_end, col_start:col_end]
            cond_num = np.linalg.cond(B)
            results[case]['cond'].append(cond_num)

            # Print sample data for small m values
            if m <= 5 and case == 'N=4m+3':  # Just print one case for brevity
                print(f"{case}: m={m}, N={N}, B shape={B.shape}, Cond={cond_num:.2f}")

    # Plot all cases
    plt.figure(figsize=(12, 7))
    for case, config in cases.items():
        plt.plot(m_values, results[case]['cond'],
                 marker=config['marker'],
                 color=config['color'],
                 linestyle='-',
                 linewidth=2,
                 markersize=8,
                 label=case)

    plt.xlabel('m', fontsize=12)
    plt.ylabel('Condition Number (log scale)', fontsize=12)
    plt.title('Comparative Analysis of Condition Numbers\nfor Different N Cases (m=2 to 20)',
              fontsize=14, pad=20)
    plt.yscale('log')  # Log scale for better visualization
    plt.grid(True, linestyle='--', alpha=0.7, which='both')
    plt.xticks(m_values)
    plt.legend(fontsize=12, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    return results

# Analyze for m values from 2 to 20
m_values = range(2, 21)
results = analyze_all_cases(m_values)
```
