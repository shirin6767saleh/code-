```python
import numpy as np

m = 2
N = 4 * m

# Loop over M from 0 to N-1
for M in range(N):
    W = np.exp((-2j * np.pi) / N)  # base complex exponential
    H = np.zeros((N, N), dtype=complex)

    # Build matrix H
    for k in range(N):
        for l in range(N):
            exponent = (k - (M + 1) / 2) * (l - (M + 1) / 2)
            H[k, l] = (1 / np.sqrt(N)) * (W ** exponent)

    # Calculate the second and fourth powers of H
    H_squared = np.linalg.matrix_power(H, 2)
    H_fourth = np.linalg.matrix_power(H, 4)

    # Print the results
    print(f"Results for M = {M}:")
    print("H^2 (Second power):")
    print(H_squared)
    print("H^4 (Fourth power):")
    print(H_fourth)
    print("-" * 40)
```
